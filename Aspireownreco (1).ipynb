```python
def convert_to_list(text):
    return text.split(',')
```

```python
pre_careers['Subjects in Schools'] = pre_careers['Subjects in Schools'].apply(convert_to_list)
```

```python
pre_careers['Required Skills']=pre_careers['Required Skills'].apply(convert_to_list)
pre_careers['Degrees to Pursue']=pre_careers['Degrees to Pursue'].apply(convert_to_list)
pre_careers.head()
```

```python
pre_careers['Degrees to Pursue']
```

```python
df_backup = pre_careers.copy()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
pre_careers['Required Skills']=pre_careers['Required Skills'].apply(collapse)
pre_careers['Subjects in Schools']=pre_careers['Subjects in Schools'].apply(collapse)
pre_careers.head()
```

```python
pre_careers = pre_careers[['Career Path','Details','Required Skills','Degrees to Pursue','Subjects in Schools','Market Trends','Image']]
pre_careers.head()
```

```python
pre_careers.shape
```

```python
pre_careers['Career Tags'] = pre_careers['Subjects in Schools'] + pre_careers['Required Skills']
```

```python
pre_careers.head()
```

```python
careersdff = pre_careers[['Career Path','Career Tags','Degrees to Pursue','Market Trends','Image','Details']]
careersdff.head()
```

```python
careersdff['Career Tags'][0]
```

```python
def remove_spaces_in_place(str_list):
    for i in range(len(str_list)):
        str_list[i] = str_list[i].replace(" ", "")
careersdff['Career Tags'].apply(remove_spaces_in_place)
careersdff.head()
```

```python
careersdff['Career Tags'] = careersdff['Career Tags'].apply(lambda x: " ".join(x))
careersdff.head()
```

```python
careersdff['Career Tags'] = careersdff['Career Tags'].apply(lambda x: x.lower())
careersdff.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv1 = CountVectorizer(max_features=200,stop_words='english')
```

```python
careersdff['Career Tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
careersdff['Career Tags'] = careersdff['Career Tags'].apply(stem)
careersdff.head()
```

```python
careerdf = careersdff.copy()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector = cv1.fit_transform(careersdff['Career Tags']).toarray()
cv1.get_feature_names_out()
vector
```

```python
inpu = 'economics'
inpu_stemmed = stem(inpu)
```

```python
inpu_vectorized = cv1.transform([inpu_stemmed]).toarray()
similarity = cosine_similarity(inpu_vectorized,vector)
similarity
```

```python
def recommend(inpui):
    inpu_stemmed = stem(inpui)
    inpu_vectorized = cv1.transform([inpu_stemmed]).toarray()
    similarity = cosine_similarity(inpu_vectorized,vector)
    distances = sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])
    for i in distances[0:5]:
        print(careersdff.iloc[i[0]]['Career Path'])
```

```python
duplicates = careersdff[careersdff.duplicated(subset='Career Path', keep=False)]
duplicates
```

```python
recommend('biology chemistry physics mathematics')
```
